## **Capturar variables**

```python
# Capturar los tipos de variables
import numpy as np
import pandas as pd
from Funciones import downcast_dtypes, nan_values

data = pd.read_csv('https://raw.githubusercontent.com/jorge-robledo11/Datasets/main/Datasets/houseprice.csv')
data.head()
```

```python
data = downcast_dtypes(data)
```

```python
import io

buffer = io.StringIO()
data.info(buf=buffer, memory_usage='deep')
s = buffer.getvalue()

with open('data_info.txt', 'w', encoding='utf-8') as f:  
    f.write(s)
```

```python
# Datetime variables
datetime_predictores = ['GarageYrBlt','YearRemodAdd','YrSold','YearBuilt']

# for i in data[datetime_predictores]:
#     data[i] = pd.to_datetime(data[i])

data[datetime_predictores] = data[datetime_predictores].apply(lambda x: pd.to_datetime(x))

# Dropeamos variables que no sirven
data.drop(['Id','SalePrice'], inplace=True, axis=1)
```

```python
# Capturar los tipos de variables
numericals = list(data.select_dtypes(include = [np.int16, np.float32]).columns)
categoricals = list(data.select_dtypes(include = ['category', 'bool']).columns)
temporary = list(data.select_dtypes(include = ['datetime', 'timedelta']).columns)
discretes = [col for col in data[numericals] if len(data[numericals][col].unique()) < 20]
continuous = [col for col in data[numericals] if col not in discretes]

# Variables
print(f'Hay {len(numericals)} variables numÃ©ricas')
print(f'Hay {len(continuous)} variables continuas')
print(f'Hay {len(discretes)} variables discretas')
print(f'Hay {len(temporary)} variables temporales')
print(f'Hay {len(categoricals)} variables categÃ³ricas')
```

```python
# Valores faltantes
nan_values(data=data, variables=continuous, variable_type='Continuas')
print()
nan_values(data=data, variables=discretes, variable_type='Discretas')
print()
nan_values(data=data, variables=temporary, variable_type='Temporales')
print()
nan_values(data=data, variables=categoricals, variable_type='CategÃ³ricas')
```

```python
# Downcasting
from pandas import DataFrame

def downcast_dtypes(dataframe: DataFrame) -> DataFrame:

    start = dataframe.memory_usage(deep=True).sum() / 1024 ** 2
    
    float_cols = [col for col in dataframe if dataframe[col].dtype == 'float64']
    int_cols = [col for col in dataframe if dataframe[col].dtype in ['int64', 'int32']]
    object_cols = [col for col in dataframe if dataframe[col].dtype in ['object', 'bool']]

    dataframe[float_cols] = dataframe[float_cols].astype(np.float32)
    dataframe[int_cols] = dataframe[int_cols].astype(np.int16)
    dataframe[object_cols] = dataframe[object_cols].astype('category')

    end = dataframe.memory_usage(deep=True).sum() / 1024 ** 2
    saved = (start - end) / start * 100
    print(f'Memory Saved: {saved:0.2f}%', '\n')
    dataframe.info()
    
    return dataframe
```
