# **Estadística Descriptiva**

```python
# Librerías
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Dataset
heladería = pd.DataFrame({
    'Nueva York': [12327],
    'Los Ángeles': [17129],
    'San Francisco': [19923]
})

# Barplot
plt.style.use('dark_background')
plt.title('Barplot')
sns.barplot(data=heladería, palette='Reds_d', saturation=0.75)
plt.grid()
plt.tight_layout()
```

```python
# Barplot horizontal
plt.style.use('seaborn-pastel')
heladería.plot.barh(color=['red', 'blue', 'green'])
plt.grid()
```

```python
# Piechart
total = heladería['San Francisco'].sum() + heladería['Nueva York'].sum() + heladería['Los Ángeles'].sum()
print(f'Total: {total:0.2f}')

ny_frecuencia = float(heladería['Nueva York'] / total)
print(f'Nueva York: {ny_frecuencia:0.2f}')

la_frecuencia = float(heladería['Los Ángeles'] / total)
print(f'Los Ángeles: {la_frecuencia:0.2f}')

sf_frecuencia = float(heladería['San Francisco'] / total)
print(f'San Francisco: {sf_frecuencia:0.2f}')

plt.title('Piechart')
plt.style.use('seaborn-colorblind')
plt.pie(data=heladería, x=[ny_frecuencia, la_frecuencia, sf_frecuencia], colors=['teal', 'indigo', 'brown'],
        labels=heladería.columns, autopct = '%0.2f%%', shadow=True)
plt.legend(loc='lower left')
plt.tight_layout()
```

## **Desviación estándar y Coeficiente de variación**

```python
# Dataset
ingreso_mensual.rename(columns={'Ingreso Mensual': 'USA'}, inplace=True)
usa = ingreso_mensual.copy()
usa.head()
```

```python
# Dataset
valores = [62000, 63000, 76000, 79000, 67000, 66000, 69000, 68000, 57000, 70000, 67000]
valores = {'Dinamarca': valores}
dinamarca = pd.DataFrame(valores)
dinamarca.head()
```

```python
# Estimadores
print('Estimadores')

# Media
media_usa = usa.mean()
print(f'• USA media: {float(media_usa):0.2f}')

media_dinamarca = dinamarca.mean()
print(f'• Dinamarca media: {float(media_dinamarca):0.2f}')

print('-'*36)

# Varianza
varianza_usa = usa.var()
print(f'• USA varianza: {float(varianza_usa):0.2f}')

varianza_dinamarca = dinamarca.var()
print(f'• Dinamarca varianza: {float(varianza_dinamarca):0.2f}')

print('-'*40)

# Desviación estándar
std_usa = usa.std()
print(f'• USA desviación estándar: {float(std_usa):0.2f}')

std_dinamarca = dinamarca.std()
print(f'• Dinamarca desviación estándar: {float(std_dinamarca):0.2f}')

print('-'*42)

# Coeficiente de variación
coeficiente_variacion_usa = std_usa / media_usa
print(f'• Coeficiente variación USA: {float(coeficiente_variacion_usa):0.2f}')

coeficiente_variacion_dinamarca = std_dinamarca / media_dinamarca
print(f'• Coeficiente variación Dinamarca: {float(coeficiente_variacion_dinamarca):0.2f}')

# Razonamiento
print("""\nRazonamiento
• Dinamarca es un país mucho más igualitario que los Estados Unidos
• No sólo la varianza es menor, sino también la desviación estándar de los salarios
• Puedes tener la sensación de que casi toda la gente del país gravita en torno a los mismos ingresos""")
```

## **Correlación**

```python
# Dataset
calificaciones = pd.read_csv('D:\Python Scripts & Notebooks\Jupyter Notebooks\Artificial Intelligence\Machine Learning\Templates\lectura y escritura.csv')
calificaciones.head()
```

```python
# Muestra
muestra = len(calificaciones)
print(f'• Muestra: {muestra}')

# Media
escritura_media = calificaciones['Escritura'].mean()
print(f'• Escritura media: {escritura_media:0.2f}')

lectura_media = calificaciones['Lectura'].mean()
print(f'• Lectura media: {lectura_media:0.2f}')

# Suma de variaciones
def suma(x:DataFrame, y:DataFrame):
    return (x - escritura_media) * (y - lectura_media)

suma = suma(x=calificaciones['Escritura'], y=calificaciones['Lectura'])
suma = sum(suma)
print(f'• Suma: {suma:0.2f}')

# Correlación
correlacion = calificaciones['Escritura'].corr(calificaciones['Lectura'])
print(f'• Coeficiente de correlación {correlacion:0.2f}')
```
