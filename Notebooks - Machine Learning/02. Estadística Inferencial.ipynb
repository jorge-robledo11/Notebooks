# **Estadística Inferencial**

```python
# Librerías
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## **Intervalos de confianza**
### **Desviación estándar poblacional conocida**

```python
# Dataset
salaries = pd.read_excel('D:\Python Scripts & Notebooks\Jupyter Notebooks\Artificial Intelligence\Machine Learning\Templates\intervalos de confianza.xlsx')
salaries.head()
```

```python
# Estimadores
print('Estimadores')

# Media
media = salaries['Conjunto de datos'].mean()
print(f'• Media: {media:0.2f}')

# Desviación estándar poblacional
std = 15000
print((f'• Desviacón estándar: {std:0.2f}'))

# Error estándar
standard_error = std / np.sqrt(len(salaries))
print((f'• Error estándar: {standard_error:0.2f}'))

# Valor crítico
z_score = 1.65
print(f'• z-score: {z_score}')

# Intervalos de confianza resultado
ic_bajo = media - z_score * (std/np.sqrt(len(salaries)))
print((f'• IC bajo: {ic_bajo:0.2f}'))

ic_alto = media + z_score * (std/np.sqrt(len(salaries)))
print((f'• IC alto: {ic_alto:0.2f}'))

# Razonamiento
print(f'\nEstamos un 90% seguros que el salario promedio se encuentra en el intervalo: [${ic_bajo:0.2f};${ic_alto:0.2f}] dólares')
```

```python
# Dataset
salaries2 = pd.read_excel('D:\Python Scripts & Notebooks\Jupyter Notebooks\Artificial Intelligence\Machine Learning\Templates\distribucion t-student.xlsx')
salaries2.head()
```

```python
# Estimadores
print('Estimadores')

# Media muestral
media = salaries2['dataset'].mean()
print(f'• Media: {media:0.2f}')

# Desviación estándar
std = salaries2['dataset'].std()
print(f'• Desviación estándar: {std:0.2f}')

# Error estándar
standard_error = std / np.sqrt(len(salaries2))
print((f'• Error estándar: {standard_error:0.2f}'))

# Valor crítico
t_score = 3.355
print(f'• t-score: {t_score}')

# Intervalos de confianza resultado
ic_bajo = media - t_score * (std/np.sqrt(len(salaries2)))
print((f'• IC bajo: {ic_bajo:0.2f}'))

ic_alto = media + t_score * (std/np.sqrt(len(salaries2)))
print((f'• IC alto: {ic_alto:0.2f}'))

# Razonamiento
print("""\nRazonamiento
• La varianza poblacional es desconocida
• Tenemos una muestra pequeña
• Suponemos que la población se distribuye en forma normal
• La estadística apropiada a utilizar es la estadística t-Student""")
print(f'\nEstamos un 99% seguros que el salario promedio se encuentra en el intervalo: [${ic_bajo:0.2f};${ic_alto:0.2f}] dólares')
```

### **Muestras independientes**
### **Varianza poblacional desconocida pero asumida como iguales**

```python
# Dataset
fruits = pd.read_csv('D:\Python Scripts & Notebooks\Jupyter Notebooks\Artificial Intelligence\Machine Learning\Templates\data fruits.csv')
fruits.tail()
```

```python
# Estimadores
print('Estimadores')

# Media
ny_media = fruits['Manzanas en NY'].mean()
print(f'• Media NY: {ny_media:0.2f}')

la_media = fruits['Manzanas en LA'].mean()
print(f'• Media LA: {la_media:0.2f}')
print('-'*50)

# Desviación estándar
ny_std = fruits['Manzanas en NY'].std()
print(f'• Desviación estándar NY: {ny_std:0.2f}')

la_std = fruits['Manzanas en LA'].std()
print(f'• Desviación estándar LA: {la_std:0.2f}')
print('-'*50)

# Varianza y Desviación combinada
# Función para hallar la varianza muestral combinada
def varianza_muestral_combinada(data:DataFrame, data2:DataFrame, std:float, std2:float):
    tmp = (len(data) - 1) * std**2 + (len(data2) - 1) * std2**2
    tmp = tmp / (len(data) + len(data2) - 2)
    return tmp

vmc = varianza_muestral_combinada(data=fruits['Manzanas en NY'], data2=fruits['Manzanas en LA'], std=ny_std, std2=la_std)

print(f'• Varianza Muestral Combinada: {vmc:0.2f}')
print(f'• Desviacón estándar Muestral Combinada: {np.sqrt(vmc):0.2f}')
print('-'*50)

# Valor Crítico
t_score = 1.746
print(f'• t-score: {t_score}')
print('-'*50)

# Intervalos de confianza resultado
ic_bajo = (ny_media - la_media) - t_score * np.sqrt(vmc/len(fruits['Manzanas en NY']) + vmc/len(fruits['Manzanas en LA']))
print((f'• IC bajo con un 90% de confianza: {ic_bajo:0.2f}'))

ic_alto = (ny_media - la_media) + t_score * np.sqrt(vmc/len(fruits['Manzanas en NY']) + vmc/len(fruits['Manzanas en LA']))
print((f'• IC alto con un 90% de confianza: {ic_alto:0.2f}'))
print('-'*120)

# Razonamiento
print(f'Tenemos un 90% de confianza que el valor promedio de las manzanas se encuentra en el intervalo: [${ic_bajo:0.2f};${ic_alto:0.2f}] dólares')
print(f'Por ende las manzanas en NY son mucho más costosas que en LA')
```
